```python
#@title A. Google Drive Mount & 기타 환경 구성
#@markdown 세모 버튼을 눌러 실행하세요!(^w^) 드라이브 마운트를 위해 허용 요청이 발생합니다.<br>훈련에 사용될 'MiriVoice'폴더와 하위 폴더들이 드라이브에 생성됩니다.(해당 폴더들이 이미 존재할 경우 생성하지 않고 넘어갑니다.)<br><br>드라이브의 MIriVoice 관련 폴더들을 완전히 삭제하고 다시 시작하고 싶을 경우, 하단의 **remove_mirivoice_directory**에 체크한 후 셀을 다시 실행하세요.<br>if you cannot understand Korean, plz **View Code**. There's comments line by line in English<br>**PLZ CHECK: GPU가 활성화되어 있는지 꼭 확인하세요.**.
#when you delete current MIriVoice Folder↓
remove_mirivoice_directory = False #@param {type: "boolean"} 

#import modules↓
from google.colab import drive
import os
from shutil import rmtree
mirivoice_directory_path = '/content/drive/MyDrive/MiriVoice'

!echo "*** Start"
#mount to google drive↓
!echo '>> mounting at google drive...'
drive.mount('/content/drive', force_remount= 1)
!echo

#make MiriVoice directories↓
if remove_mirivoice_directory: #←when you delete and reset MIriVoice directory.
  rmtree(mirivoice_directory_path)
  #remove↓
  !echo '>> Removed {mirivoice_directory_path} successfully.'
  
  #remake↓
  os.makedirs(mirivoice_directory_path)
  !echo '>> Remade {mirivoice_directory_path} successfully.'
  !echo

  os.makedirs(f'{mirivoice_directory_path}/PretrainedModels')
  !echo '>> Remade {mirivoice_directory_path}/PretrainedModels successfully.'
  
  os.makedirs(f'{mirivoice_directory_path}/Dataset/WavsAndLabs')
  !echo '>> Remade {mirivoice_directory_path}/Dataset/WavsAndLabs successfully.'
  
  os.makedirs(f'{mirivoice_directory_path}/Dataset/TextGrids')
  !echo '>> Remade {mirivoice_directory_path}/Dataset/TextGrids successfully.'

  os.makedirs(f'{mirivoice_directory_path}/Dataset/PreprocessedDatas')
  !echo '>> Remade {mirivoice_directory_path}/Dataset/PreprocessedDatas successfully.'
  
  os.makedirs(f'{mirivoice_directory_path}/Results/CheckPoints')
  !echo '>> Remade {mirivoice_directory_path}/Results/CheckPoints successfully.'
  
  os.makedirs(f'{mirivoice_directory_path}/Results/Log')
  !echo '>> Remade {mirivoice_directory_path}/Results/Log successfully.'
  
  os.makedirs(f'{mirivoice_directory_path}/Results/Evaluate')
  !echo '>> Remade {mirivoice_directory_path}/Results/Evaluate successfully.'
  
  os.makedirs(f'{mirivoice_directory_path}/Results/Test')
  !echo '>> Remade {mirivoice_directory_path}/Results/Test successfully.'
else: #←when you not delete MIriVoice directory.
  #make↓
  os.makedirs(mirivoice_directory_path, exist_ok = True)
  !echo '>> {mirivoice_directory_path} exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/PretrainedModels', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/PretrainedModels exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/Dataset/WavsAndLabs', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/Dataset/WavsAndLabs exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/Dataset/TextGrids', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/Dataset/TextGrids exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/Dataset/PreprocessedDatas', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/Dataset/PreprocessedDatas exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/Results/CheckPoints', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/Results/CheckPoints exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/Results/Log', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/Results/Log exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/Results/Evaluate', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/Results/Evaluate exists currently.'

  os.makedirs(f'{mirivoice_directory_path}/Results/Test', exist_ok = True)
  !echo '>> {mirivoice_directory_path}/Results/Test exists currently.'
!echo

#clone from github to train tts models↓
!echo '>> cloning repository from github...'
!git clone 'https://github.com/EX3exp/TrainForMiriVoice.git'
!echo '*** Finish'
```

```python
#@title B. 데이터 전처리를 위한 기본 설정
!echo "*** Start"

#@markdown 1. 데이터셋의 이름을 입력합니다. <br> custom_dataset_name은 커스텀한 Corpus를 사용하는 경우에만 작성합니다. <br>**커스텀한 Corpus를 사용하는 경우, 반드시 is_customed_corpus에 체크해 주세요.**<br>실제 코드의 변수엔 하단의 두 입력값 중 하나만 반영되므로, 걱정 마시고 해당하는 양식만 채워 주시면 됩니다.
#Check this box if you used your customed-corpus in your dataset↓
is_customed_corpus = True #@param {type: "boolean"}

#define dataset's name↓
if is_customed_corpus: #←when you use your own-korean-corpus.
  customed_dataset_name = "sce-tts"#@param {type: "string"}
  dataset_name = customed_dataset_name
else:#←when you use korean-corpus which currently supported in MiriVoice.
  embeded_dataset_name = "EX3Corpus"#@param ["EX3Corpus"] {type: "raw"}
  dataset_name = embeded_dataset_name

#@markdown <br> 3. metadata가 적힌 .txt의 파일명을 입력합니다. <br> ※주의: 이 파일은 **데이터셋 폴더의 바로 아래에 위치해 있어야 하며, 다음의 형식을 갖추어야 합니다.**<br>SCE0001.wav|우리는 쉑쉑버거를 먹으러 왔어요.<br>SCE0002.wav|습격을 받았을 때 다친 상처 때문에 동굴에 인질로 갇혀 있던 과학자의 도움을 받습니다.<br>SCE0003.wav|당신의 아버지는 부엌에서 요리합니까?
#filename of your metadata.↓
metadata_name= "metadata.txt"#@param {type: "string"}

#@markdown <br> 5. 데이터셋에 있는 .wav들의 샘플링레이트를 입력합니다.
#sampling rate of your dataset's wavs.↓
_sampling_rate = 22050 #@param {type: "integer"} 

#@markdown <br> 6.사전학습된 FastSpeech2 Model의 이름을 입력합니다.
#filename of your pretrained FastSpeech2 Model.↓ 
pretrained_fastspeech2_name = "checkpoint_350000.pth.tar"#@param {type: "string"}

#@markdown <br> 7.사전학습된 Vocoder Model의 종류와 이름을 입력합니다. 
#type of your pretrained vocoder Model.↓
pretrained_vocoder_type = "vocgan"#@param ["vocgan"]
#filename of your pretrained vocoder Model.↓
pretrained_vocoder_name = "vocgan_kss_pretrained_model_epoch_4500.pt"#@param {type: "string"}

#// write hparam.py // we will use this file at training process. ↓
hparam = f'''import os
dataset = "{dataset_name}"
data_path = "{mirivoice_directory_path}/Dataset/"
meta_name = "{metadata_name}"
textgrid_path= "{mirivoice_directory_path}/Dataset/TextGrids/"
textgrid_name = "TextGrids.zip"

### set GPU number ###
train_visible_devices = "0,1"
synth_visible_devices = "0"

# Text
text_cleaners = ['korean_cleaners']

# Audio and mel
sampling_rate = {_sampling_rate}
filter_length = 1024
hop_length = 256
win_length = 1024

max_wav_value = 32768.0
n_mel_channels = 80
mel_fmin = 0
mel_fmax = 8000

f0_min = 71.0
f0_max = 792.8
energy_min = 0.0
energy_max = 283.72


# FastSpeech 2
encoder_layer = 4
encoder_head = 2
encoder_hidden = 256
decoder_layer = 4
decoder_head = 2
decoder_hidden = 256
fft_conv1d_filter_size = 1024
fft_conv1d_kernel_size = (9, 1)
encoder_dropout = 0.2
decoder_dropout = 0.2

variance_predictor_filter_size = 256
variance_predictor_kernel_size = 3
variance_predictor_dropout = 0.5

max_seq_len = 1000

# Checkpoints and synthesis path
preprocessed_path = "{mirivoice_directory_path}/Dataset/PreprocessedDatas"
checkpoint_path = "{mirivoice_directory_path}/Dataset/Results/CheckPoints"
eval_path = "{mirivoice_directory_path}/Dataset/Results/Evaluate"
log_path = "{mirivoice_directory_path}/Dataset/Results/Log"
test_path = "{mirivoice_directory_path}/Dataset/Results/Test"


# Optimizer
batch_size = 4
epochs = 1000
n_warm_up_step = 4000
grad_clip_thresh = 1.0
acc_steps = 1

betas = (0.9, 0.98)
eps = 1e-9
weight_decay = 0.

#pretrained Model
pretrained_model_name = "{pretrained_fastspeech2_name}"
pretrained_model_path = "{mirivoice_directory_path}/PretrainedModels/{pretrained_fastspeech2_name}"

# Vocoder
vocoder = "{pretrained_vocoder_type}"
vocoder_pretrained_model_name = "{pretrained_vocoder_name}"
vocoder_pretrained_model_path = "{mirivoice_directory_path}/PretrainedModels/{pretrained_vocoder_name}"

# Log-scaled duration
log_offset = 1.


# Save, log and synthesis
save_step = 10000
eval_step = 1000
eval_size = 256
log_step = 1000
clear_Time = 20'''

with open("/content/TrainForMiriVoice/hparams.py", "w") as f:
  f.write(hparam)

!echo '>> Modified /content/TrainForMiriVoice/hparams.py successfully.'
!echo

#function that gets value from hparams.py↓
def get_value_from_hparams(value):
  exec(hparam)
  _value = value
  return _value

!echo '*** Finish'
```

```python
#@title D. 데이터 전처리 (Data Preprocess)
#@markdown 세모 버튼을 눌러 데이터 전처리를 시작합니다. 실행이 끝날 때까지 기다려 주세요.<br>정상적으로 진행되었을 경우, PreprocessedDatas 폴더 내에 파일들이 생성되어 있을 것입니다.
!echo "*** Start"

while 'Errors not found':
  !echo ">> Checking necessary files before starting process..."
  !echo
  if os.path.isfile(f'{mirivoice_directory_path}/Dataset/train.zip'):
    !echo '>> {mirivoice_directory_path}/Dataset/train.zip exists currently.'
    train_zip_not_exists = False
  else:
    !echo ">> Plz put train.zip in {mirivoice_directory_path}/Dataset/."
    train_zip_not_exists = True

  if os.path.isfile(f'{mirivoice_directory_path}/Dataset/val.zip'):
    !echo '>> {mirivoice_directory_path}/Dataset/val.zip exists currently.'
    val_zip_not_exists = False
  else:
    !echo "\n>> Plz put val.zip in {mirivoice_directory_path}/Dataset/."
    val_zip_not_exists = True

  if train_zip_not_exists and val_zip_not_exists:
    !echo
    !echo ">> Notice: train.zip and val.zip not exists in {mirivoice_directory_path}/Dataset/. We can't start process now."
    !echo '*** Unexpected exit'
    break

  !echo
  !echo '>> unzipping val.zip'
  %cd '{mirivoice_directory_path}/Dataset/WavsAndLabs'
  !unzip -qq '{mirivoice_directory_path}/Dataset/val.zip'
  
  !echo
  !echo '>> unzipping train.zip'
  %cd '{mirivoice_directory_path}/Dataset/WavsAndLabs'
  !unzip -qq '{mirivoice_directory_path}/Dataset/train.zip'

  %cd '/content/TrainForMiriVoice'

  #Runs preprocess.py↓
  !echo 'Running preprocess.py...'
  !echo
  !python '/content/TrainForMiriVoice/preprocess.py'
  !echo '>> If errors not occured: data preprocessed successfully.' 
  !echo
  !echo "*** Finish"
  break
```
